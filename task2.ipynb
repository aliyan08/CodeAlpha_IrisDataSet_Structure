```python
from sklearn.datasets import load_iris

data = load_iris()
X = data.data
y = data.target
```

```python
import pandas as pd
df = pd.DataFrame(X, columns=data.feature_names)
df['target'] = y
```

```python
df
```

```python
df.isna().sum()
```

```python
# visualize the data
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(df, hue='target')
plt.show()
```

```python
# apply K-means and visualize the result
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
df['cluster'] = kmeans.labels_
sns.pairplot(df, hue='cluster')
plt.show()
```

```python
# 3D visualization
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['sepal length (cm)'], df['sepal width (cm)'], df['petal length (cm)'], c=df['cluster'], cmap='rainbow')
ax.set_xlabel('sepal length (cm)')
ax.set_ylabel('sepal width (cm)')
ax.set_zlabel('petal length (cm)')
plt.show()
```

```python
# scatter plot
fig = plt.figure()
ax = fig.add_subplot()
ax.scatter(df['sepal length (cm)'], df['sepal width (cm)'], df['petal length (cm)'], c=df['target'], cmap='rainbow')
ax.set_xlabel('sepal length (cm)')
ax.set_ylabel('sepal width (cm)')
plt.show()
```
